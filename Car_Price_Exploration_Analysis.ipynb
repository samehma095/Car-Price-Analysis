```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car_price_dataset_medium.csv")
```

```python
df.dropna()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Brand', order=df['Brand'].value_counts().index, palette='viridis')
plt.title('Most Frequent Car Brands')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Fuel_Type', y='Price_USD', hue='Transmission')
plt.title('Price by Fuel Type and Transmission')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Max_Power_bhp', y='Price_USD', hue='Fuel_Type', alpha=0.7)
plt.title('Relationship Between Engine Power and Price')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# بنختار الأعمدة الرقمية فقط عشان نحسب الارتباط
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Analysis: What affects the Price?')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
df.groupby('Brand')['Price_USD'].mean().sort_values(ascending=False).plot(kind='bar', color='black')
plt.title('Average Car Price by Brand')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Kilometers_Driven', y='Price_USD', alpha=0.5)
plt.title('Impact of Mileage on Car Price')
plt.show()
```

```python
print(f"Total Cars Analyzed: {len(df)}")
print(f"Average Car Price: ${df['Price_USD'].mean():.2f}")
print(f"Most Expensive Brand: {df.groupby('Brand')['Price_USD'].mean().idxmax()}")
print(f"Average Engine CC: {df['Engine_CC'].mean():.0f}")
```
